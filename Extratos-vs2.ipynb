```python
#Código Completo

#Bibliotecas
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
import openpyxl
import pyautogui

#Abrindo o Navegador
servico = Service(ChromeDriverManager().install())
navegador =  webdriver.Chrome(service=servico)

#Acessar no Newcon Web
navegador.get("https://newcon.bancorbras.com.br/intranet/")
time.sleep(2)
#Usuário
navegador.find_element(By.ID, 'edtUsuario').send_keys('USUÁRIO')
time.sleep(2)
#Senha
navegador.find_element(By.ID, 'edtSenha').send_keys('SENHA')
time.sleep(2)
#Login
navegador.find_element(By.ID, 'btnLogin').click()
time.sleep(2)
#Unidade de Negócio
navegador.find_element(By.XPATH, '//*[@id="rptUnidadeNegocio_ctl00_txt_CD_Unidade_Negocio"]').click()
time.sleep(2)

pasta_saida = 'extratos_gerados'

#Importanto a planilha.
workbook = openpyxl.load_workbook('extrato.xlsx')
#Indicando qual aba eu quero utilizar.
pagina_clientes = workbook['Plan1']
#Informando qual a primeira linha de leitura
for linha in pagina_clientes.iter_rows(min_row=2):
    grupo = linha [0].value
    cota = linha[1].value
    nome = linha[2].value

    #Atendimento
    navegador.find_element(By.ID, 'ctl00_img_Atendimento').click()
    time.sleep(2)
    #Gupo
    navegador.find_element(By.XPATH, '//*[@id="ctl00_Conteudo_edtGrupo"]').send_keys(str(grupo))
    time.sleep(2)
    #Cota
    navegador.find_element(By.XPATH, '//*[@id="ctl00_Conteudo_edtCota"]').click()
    time.sleep(2)
    navegador.find_element(By.XPATH, '//*[@id="ctl00_Conteudo_edtCota"]').send_keys(str(cota))
    time.sleep(2)
    #Localizar
    navegador.find_element(By.ID, 'ctl00_Conteudo_btnLocalizar').click()
    time.sleep(2)
    #Imprimir extrato
    navegador.find_element(By.XPATH, '//*[@id="ctl00_Conteudo_Menu_CONAT_grdMenu_CONAT_ctl19_hlkFormulario"]').click()
    time.sleep(2)
    #Desmarcar Mensagem
    navegador.find_element(By.XPATH, '//*[@id="ctl00_Conteudo_cblOpcoesExtrato_17"]').click()
    time.sleep(2)
    #Imprimir
    navegador.find_element(By.XPATH, '//*[@id="ctl00_Conteudo_btnImprimir"]').click()
    time.sleep(2)
    #Extrato Atualizado
    navegador.find_element(By.XPATH, '//*[@id="ctl00_Conteudo_grdHistoricoSituacao"]/tbody/tr[2]').click()
    time.sleep(4)

    #Imprimir
    pyautogui.hotkey('win', 'up')
    time.sleep(2)
    x, y, largura, altura = pyautogui.locateOnScreen('imprimir-2.png', grayscale= True, confidence= 0.7)
    time.sleep(1)
    pyautogui.click(x + largura/2, y + altura/2)
    time.sleep(2)
    pyautogui.press('enter')
    time.sleep(4)
    #Renomear Arquivo
    pyautogui.write(f'{grupo}-{cota} - Extrato 10-07-20225')
    time.sleep(1)
    pyautogui.press('enter')
    time.sleep(4)

    #Fechar
    pyautogui.hotkey('alt', 'f4')
    time.sleep(2)

    #Fechar
    navegador.find_element(By.XPATH, '//*[@id="ctl00_Conteudo_lkbfecha_historico"]').click()
    time.sleep(2) 
```
